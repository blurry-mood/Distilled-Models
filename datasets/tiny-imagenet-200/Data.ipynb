```python
from glob import glob

import pandas as pd
```

```python
train = pd.DataFrame()
```

* Read train paths
* Extract code from path

```python
paths = glob('*/train/*/images/*')
```

```python
train['img_path'] = paths
train
```

```python
train['code'] = train.img_path.apply(lambda x: x.split('/')[2])
train
```

* Read val images paths
* Extract code from val_annots.txt

```python
val = pd.DataFrame()
paths = glob('*/val/images/*')
annotations = open('tiny-imagenet-200/val/val_annotations.txt', 'r').read().split('\n')
```

```python
del annotations[-1]
codes = {k.split('\t')[0]:k.split('\t')[1] for k in annotations}
```

```python
val['img_path'] = paths
val
```

```python
val['code'] = val.img_path.apply(lambda x: codes[x.split('/')[-1]])
val
```

* Concatenate train & val

```python
data = pd.concat([train, val]).reset_index(drop=True)
data
```

* Extract label from code using words.txt

```python
words = open('tiny-imagenet-200/words.txt', 'r').read().split('\n')
```

```python
code_label = codes = {k.split('\t')[0]:k.split('\t')[1] for k in words}
```

```python
data['labels'] = data.code.apply(lambda x: code_label[x])
```

```python
data.drop('code', axis=1, inplace=True)
data
```

* Save dataframe to csv

```python
data.to_csv('data.csv', index=False)
```
